```python
from selenium import webdriver
import pandas as pd
```

```python
driver=webdriver.Chrome('C:\Program Files (x86)\chromedriver.exe')

driver.get('https://collegedunia.com/india-colleges')
```

```python
colleges=driver.find_elements_by_tag_name('h3')
location=driver.find_elements_by_xpath('//*[@id="__next"]/div[3]/section/div/div[2]/div[2]/div[4]/div/div/div[2]/div/span/span[1]')
fees=driver.find_elements_by_xpath('//*[@id="__next"]/div[3]/section/div/div[2]/div[2]/div[4]/div/div/div[2]/ul/li[1]/a/span[1]')

colleges_result=[]
for i in range(len(location)):
     temporary_data={'College':colleges[i].text,
                     'Location':location[i].text,
                     'Fee':fees[i].text}

     colleges_result.append(temporary_data)
```

```python
df_data=pd.DataFrame(colleges_result)
df_data
```

```python
df_data.to_excel('college_scraping_result.xlsx', index=False)
```
